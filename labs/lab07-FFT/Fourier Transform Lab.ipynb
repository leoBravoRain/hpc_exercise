### Recursive Fast Fourier Tranform

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Comparaci√≥n Time vs par_number

```python
n = np.array([1, 2, 3, 4, 5, 6, 7 ])

# ms
par_time = np.array([0.013217, 0.054535, 0.057343, 0.091662, 0.084832, 0.095778, 0.094691])

# Plotting results
fig, (ax1) = plt.subplots(figsize = (7,6), tight_layout = True)

# first elements
ax1.plot(n, par_time)

ax1.set_xlabel('Level for paralellization')
ax1.set_ylabel('Execution Time')
ax1.legend()
ax1.title.set_text('Execution time v/s Level of recursion for paralellization for FFT')
```

Del gr√°fico se tiene que para la primera divisi√≥n del input, se obtiene el menor tiempo de ejecuci√≥n, esto quiere decir que al paralelizar el input en la primera divisi√≥n de los datos (cuando se asigna impares y pares a cada llamada recursiva por primera vez), se obtiene el menor tiempo de ejecuci√≥n.

### Execution time for differentes times comparing ser v/s par implementation

```python
n = np.array([2**5, 2**10, 2**15, 2**17, 2**18, 2**20])

# ms
seq_time = np.array([0.000331, 0.007194, 0.167116, 0.686059, 1.356930, 5.691087])
par_time = np.array([0.000360, 0.009113, 0.013338, 0.033905, 0.067120, 0.208516])

# speed up (seq time / parallel time)
speedup = seq_time/par_time

# Plotting results
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (7,6), tight_layout = True)

final = 6

# plotting execution time
ax2.plot(n[0:final], seq_time[0:final], label = 'Seq recursion')
ax2.plot(n[0:final], par_time[0:final], label = 'Par recursion')

ax2.set_xlabel('Input size')
ax2.set_ylabel('Execution Time')
ax2.legend()
ax2.title.set_text('Execution time v/s Input size for FFT')

final_first = 2

# first elements
ax1.plot(n[0:final_first], seq_time[0:final_first], label = 'Seq recursion')
ax1.plot(n[0:final_first], par_time[0:final_first], label = 'Par recursion')

ax1.set_xlabel('Input size')
ax1.set_ylabel('Execution Time')
ax1.legend()
ax1.title.set_text('Execution time v/s Input size for FFT')

# plotting speedup
ax3.plot(n[0:final], speedup[0:final])

ax3.set_xlabel('Input size')
ax3.set_ylabel('Speedup')
ax3.title.set_text('Speedup v/s Input Size in FFT')
```

Para valores peque√±os del input (menores a 1.024), la implementaci√≥n secuencial tiene menor tiempo de ejecuci√≥n que la versi√≥n paralela, debida posiblemente a que la versi√≥n paralela debe ejecutar m√°s tareas (comuncicaci√≥n, sincronizaci√≥n) por lo que para el tama√±o dado, √©sto toma mas tiempo que para la forma secuencial.

Para tama√±os mas grandes que 1.024, se tiene que en todo momento la implementaci√≥n paralela toma mucho menos tiempo que la versi√≥n secuencial.

Para el caso del speedup, este aumenta a medida que el tama√±o del input aumenta, teniendose mayores aceleraciones para mayores inputs.
